```python
# Import package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import data
input_data = pd.read_excel(r'.\GWP_PTAP_Data.xlsx', sheet_name= '10 SPDRs and S&P 500', header=1)
input_data.set_index('Date', inplace= True)
```

### (a): Compute the price return for each SPDR and the S&P 500 for each month of 2017.

> We calculate the log-return from the raw data using the numpy log function against the difference between consecutive time periods, dropping the first row to avoid NULL fields.

```python
log_return =  np.log(input_data / input_data.shift(1))  # Calculate log-return for each SPDR and S&P 500
log_return.dropna(inplace = True)
log_return #display dataframe
```

### (b) Compute the active return (relative to the S&P 500) for each SPDR for each month of 2017.

> Active returns are calculated by subtracting the log return of the S&P 500 from the log return of each SPDR for respective months.

```python
# Create sectors list
sector_list = ['XLE','XLB','XLI','XLY','XLP','XLV','XLF','XLK','XLU','XLRE']
active_returns = pd.DataFrame()
for i in sector_list:
    active_returns[i] = log_return[i] - log_return['S&P 500']
active_returns
```

### (c) Calculate the monthly tracking error for each SPDR in 2017.

> Tracking error is quickly calculated by applying numpy standard deviation to the dataframe, noting the underlying formula as supplied in course module 4 notes: Tracking Error = $\sqrt{\frac{1}{T-1} \sum_{i=1}^{T} (R_i - \bar{R})^2}$

```python
tracking_error = []   # Create empty lists

for i in sector_list:        # Loop through sector list to calculate monthly tracking error
    tracking_error = active_returns.apply(np.std)

tracking_error = pd.DataFrame(tracking_error, columns=['Tracking Error 2017'])
tracking_error_transpose = tracking_error.T     #swap rows/columns to match original data structure
tracking_error_transpose
```

**Identify the SPDR that best tracks the S&P 500 and explain how you came to this conclusion.**


> As observed in the table above, XLI (0.010241) has the lowest tracking error (standard deviation of active returns) out of the comparitive SPDRs when drawn against the returns of the benchmark S&P 500. Although not perfectly correlated, XLI exhibits a relatively small error which suggests stability in addition the close simulation of benchmark returns.

```python
# Create empty lists
tracking_err_list = []
avg_active_ret_list = []
# Loop through sector list to calculate monthly tracking error:
for i in sector_list:
    tracking_err_list.append(np.std(active_returns[i]))
    avg_active_ret_list.append(np.mean(active_returns[i]))
# Active return per tracking error
active_return_ratio_list = np.array(avg_active_ret_list)/np.array(tracking_err_list)
active_return_ratio_list

full_data = pd.DataFrame( {'Tracking Error': tracking_err_list, 'Active Return':avg_active_ret_list ,
                    'Active Return Ratio':active_return_ratio_list}, index = sector_list)

# Plot the results
plt.figure(figsize=(10,5))

for i in range(10):
    plt.scatter(full_data.iloc[i,0], full_data.iloc[i,1], label = full_data.index[i])
plt.legend()
plt.xlabel('Tracking Error')
plt.ylabel('AVG Monthly Active Return')
plt.title('Tracking Error vs AVG Montly Active Return')

plt.tight_layout()
plt.show()
```

A visual of tracking volatility against an ideal 0 active return.
