```python
# Let's start by putting the following (completely fictional) share price lists into NumPy arrays

microsoft = [120.3, 121.7, 119.8, 122.5, 123.3]
apple = [124.1, 124.3, 124.2, 124.8, 124.3]
facebook = [110.2, 112.4, 116.8, 115.3, 119.6]
```

```python
# TASK 1: Import NumPy and create arrays

# import numpy
import numpy as np
```

```python
# create numpy array by the list above
microsoft = np.array(microsoft)
apple = np.array(apple)
facebook = np.array(facebook)
```

```python
# Now let's use some NumPy functions to learn some interesting statistical information


# TASK 2A: Calculate the mean share price for each company, and store in suitably named variables.
#          Then print them out (in order - Microsoft, then Apple, then Facebook).
microsoft_mean = round(microsoft.mean(),2)
apple_mean = round(apple.mean(),2)
facebook_mean = round(facebook.mean(),2)
print("Mean values of Micsoft - Apple - Facebook:   {} - {} - {}"
      .format(microsoft_mean,apple_mean,facebook_mean))
```

```python
# TASK 2B: Calculate the standard deviation of share prices for each company using a NumPy function,
#          and store in suitably named variables. Then print them out (in order - Microsoft, then Apple, then Facebook).
microsoft_std = round(microsoft.std(),2)
apple_std = round(apple.std(),2)
facebook_std = round(facebook.std(),2)
print("Std. deviation values of Micsoft - Apple - Facebook:   {} - {} - {}"
      .format(microsoft_std,apple_std,facebook_std))
```

```python
# TASK 2C: Calculate the percentage change in share price from the first price to the last price, and store 
#          in suitably named variables. Then print them out 
#          (in order - Microsoft, then Apple, then Facebook).

# first we will create a function where it calculate the percentage change from 
# first price to last price
def price_change(np_array):
    first_price = np_array[0]
    last_price = np_array[-1]
    per_change = (last_price - first_price)/first_price * 100
    per_change = round(per_change,2)
    return per_change

microsoft_change = price_change(microsoft)
apple_change = price_change(apple)
facebook_change = price_change(facebook)

print("% change values of Micsoft - Apple - Facebook:   {} - {} - {}"
      .format(microsoft_change,apple_change,facebook_change))
```

```python
# Now we're going to load our NumPy arrays into a Pandas Dataframe
# The 5 values in each company's array is the share price for the first 5 months of 2018

# TASK 3: Import Pandas and load the arrays into a dataframe with appropriate row and column labels, and then
#         output the dataframe by simply typing its variable name on a new line

# import pandas
import pandas as pd

# first we see that the price will be the share price for the first 5 months of 2018
# Month (1,2,3,4,5)
# Our pandas table will consist of the following col: Months, (Price) Microsoft, Apple, Facebook
months = np.array([1,2,3,4,5])

price_table = pd.DataFrame(
    {'Month': months,'Microsoft': microsoft,'Apple': apple,'Facebook': facebook}).set_index('Month')
price_table
```

```python
# Finally, let's draw some interesting graphs! 
# You will need to use some of the statistical values you calculated earlier in order to do this

# TO DO: Import Matplotlib's Pyplot submodule
import matplotlib.pyplot as plt
```

```python
# TASK 4A: Draw a line graph for one of the companies, showing their share prices over time, using the dataframe.
#          Provide suitable labels. Hint: using your_dataframe_variable.loc["your_row_label", :] will extract a single row.

# In this task, we will choose Apple share price as an example.

Apple_share_price = price_table.loc[:,'Apple']

plt.xlabel("Month")
plt.ylabel("Apple Share Price")

plt.plot(Apple_share_price)

plt.show()
```

```python
# TASK 4B: Draw a bar chart, using any data structure, comparing each company's value for any statistic 
# of your choice 
#          from Task 2 (A, B or C)

# In this example we will choose the mean as the statistic comparison between Microsoft, Apple and 
# Facebook.

# Create an Numpy array for comparison
mean_compare = np.array([microsoft_mean, apple_mean, facebook_mean])
mean_label = np.array(['Microsoft', 'Apple', 'Facebook'])

plt.xlabel('Mean')
plt.ylabel('Company Name')

plt.bar(mean_label, mean_compare, alpha = 0.5)

plt.show()
```
