```python
import numpy as np
from scipy.stats import uniform
```

```python
# estimate the value of integral from 0 to Pi, sin(x)dx
np.random.seed(0)
msin_est = [None] * 50
msin_std = [None] * 50

for i in range(1,51):
    unif_array = uniform.rvs(size = i * 1000) * np.pi
    sin_val = np.sin(unif_array) * np.pi #sample from 0 to Pi for solving the problem
    msin_est[i - 1] = np.mean(sin_val) #calculate mean
    msin_std[i - 1] = np.std(sin_val)/np.sqrt(i*1000) #calculate standard deviation with the descaling
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2]*50)
plt.plot(msin_est, '.')
plt.plot(2 + np.array(msin_std) * 3)
plt.plot(2 - np.array(msin_std) * 3)
plt.show()
```

```python
from scipy.stats import norm
```

```python
# estimate the value of integral from uninfinite to uninfinite, e ^ (- x ^ 2)dx

np.random.seed(0)
sample_sizes = [1000,25000,50000]
for sample_size in sample_sizes:
    msin_est_e = [None] * int(sample_size / 1000)
    msin_std_e = [None] * int(sample_size / 1000)

    for i in range(1,int(sample_size / 1000) + 1):
        norm_array = norm.rvs(size = i * 1000)
        e_val = np.exp(-norm_array**2/2) * np.sqrt(2 * np.pi)
        msin_est_e [i - 1] = np.mean(e_val)
        msin_std_e [i - 1] = np.std(e_val) / np.sqrt(i * 1000)

    plt.plot([np.sqrt(np.pi)] * int(sample_size / 1000))
    plt.plot(msin_est_e,'.')
    plt.plot(np.sqrt(np.pi) + np.array(msin_std) * 3, 'r')
    plt.plot(np.sqrt(np.pi) - np.array(msin_std) * 3, 'r')
    plt.show()
```
